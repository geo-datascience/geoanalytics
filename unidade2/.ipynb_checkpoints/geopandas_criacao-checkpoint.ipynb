# Funções Espaciais - Criação de Geometrias


## Buffer


```python
%matplotlib inline
import os
os.environ['USE_PYGEOS'] = '0'
import geopandas as gpd

escolas_mun = gpd.read_file("../dados/escola_municipal.geojson")
escolas_par = gpd.read_file("../dados/escola_particular.geojson")
```

```python
escolas_mun['buffer']=escolas_mun['geometry'].buffer(300)
escolas_mun.head()
```

```python
escolas_mun.set_geometry('buffer', inplace=True)
escolas_mun.plot()
```

```python
escolas_par['buffer']=escolas_par['geometry'].buffer(300)
escolas_par.head()
```

```python
escolas_par.set_geometry('buffer', inplace=True)
escolas_par.plot()
```

```python
basemap = escolas_mun.plot(color='darkgreen', alpha = 0.5, figsize=(15, 15))
basemap = escolas_par.plot(ax=basemap, color='darkblue', alpha = 0.5)
```

## Convex Hull


```python
regional = gpd.read_file("../dados/REGIONAL.shp")
regional.head()
```

```python
regional["bb"] = regional["geometry"].convex_hull
```

```python
regional.set_geometry('bb', inplace=True)
basemap = regional.plot(color='blue', alpha = 0.5, figsize=(15, 15))
regional.set_geometry('geometry', inplace=True)
basemap = regional.plot(ax=basemap, color='green', edgecolor="black",alpha = 0.5)
```

## Envelope


```python
regional["bb"] = regional["geometry"].envelope
```

```python
regional.head()
```

```python
regional.set_geometry('bb', inplace=True)
basemap = regional.plot(color='blue', alpha = 0.5, figsize=(15, 15))
regional.set_geometry('geometry', inplace=True)
basemap = regional.plot(ax=basemap, color='green', edgecolor="black",alpha = 0.5)
```
