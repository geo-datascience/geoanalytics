# Recorte

Uso da função Clip:

```python
%matplotlib inline
import geopandas as gpd

escolas_par = gpd.read_file("data/escola_particular.geojson")
escolas_par['buffer']=escolas_par['geometry'].buffer(300)
escolas_par.head()
```

```python
regional = gpd.read_file("data/REGIONAL.shp")
regional.head()
```

```python
basemap = regional.boundary.plot(color='k', figsize=(10, 10))
escolas_par.plot(ax=basemap, color='red' )
```

```python
escolas_le = gpd.clip(escolas_par, regional[regional['SIGLA']=='LE'])
basemap = regional[regional['SIGLA']=='LE'].boundary.plot(color='k')
escolas_le['geometry'].plot(ax=basemap, color='red')
```
